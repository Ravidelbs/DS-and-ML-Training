```
# menghapus data yang tidak diperlukan pada data train
df_train = df_t.drop(['id', 'Keuntungan Kapital', 'Kerugian Capital'], axis=1)
df_train.head()
```

```
# menghapus data yang tidak diperlukan pada data test
df_test = df_s.drop(['id', 'Keuntungan Kapital', 'Kerugian Capital'], axis=1)
df_test.head()
```

```
# manipulasi data pada data train
obj_kp = {'?': 'Wiraswasta'}
obj_pk = {'?': 'Spesialis'}
obj_pd = {'1st-4th': 'SD', '5th-6th': 'SD', 
          '7th-8th': 'SMP', '9th': 'SMP', 
          '10th': 'SMA', '11th': 'SMA', '12th': 'SMA'}

df_train['Kelas Pekerja'] = df_train['Kelas Pekerja'].replace(obj_kp)
df_train['Pekerjaan'] = df_train['Pekerjaan'].replace(obj_pk)
df_train['Pendidikan'] = df_train['Pendidikan'].replace(obj_pd)
df_train.head()
```

```
# manipulasi data pada data test
df_test['Kelas Pekerja'] = df_test['Kelas Pekerja'].replace(obj_kp)
df_test['Pekerjaan'] = df_test['Pekerjaan'].replace(obj_pk)
df_test['Pendidikan'] = df_test['Pendidikan'].replace(obj_pd)
df_test.head()
```

```
# mengubah data kategorik menjadi numerik pada data train
obj_pdo = {'SD': 0, 'SMP': 1, 'SMA': 2, 'Pendidikan Tinggi': 3,
          'Sekolah Professional': 4, 'D3': 5, 'D4': 6, 
          'Sarjana': 7, 'Master': 8, 'Doktor': 9}
obj_gji = {'<=7jt': 0, '>7jt': 1}

df_train['Pendidikan'] = df_train['Pendidikan'].replace(obj_pdo)
df_train['Gaji'] = df_train['Gaji'].replace(obj_gji)
df_train.head()
```

```
# mengubah data kategorik menjadi numerik pada data test
df_test['Pendidikan'] = df_test['Pendidikan'].replace(obj_pdo)
df_test.head()
```

```
# mengubah tipe data pada data train
df_train['Jam per Minggu'] = df_train['Jam per Minggu'].astype(int)
df_train.head()
```

```
# mengubah tipe data pada data test
df_test['Jam per Minggu'] = df_test['Jam per Minggu'].astype(int)
df_test.head()
```

```
# memvisualisasikan data untuk melihat korelasi data
import matplotlib.pyplot as plt
import seaborn as sns

df_tv = df_train[['Umur',	'Berat Akhir', 'Pendidikan', 'Jmlh Tahun Pendidikan', 'Jam per Minggu',	'Gaji']].sample(n=10000, random_state=123)

g = sns.pairplot(df_tv, hue='Gaji', markers=['+', 'o'], hue_order=[0,1])
  
for i, j in zip (*np.triu_indices_from(g.axes, 1)):
  g.axes[i, j].set_visible(False)
```

```
# membagi data feture dan data target pada data train
x_train = df_train.drop('Gaji', axis=1)
x_train = pd.get_dummies(x_train)
y_train = df_train['Gaji']
```

```
# membagi data feature dan data predik pada data test
x_test = df_test
x_test = pd.get_dummies(x_test)
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
```

```
# membuat model dan evaluasi data dengan algoritma DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

model_clf = DecisionTreeClassifier()
model_clf.fit(x_train, y_train)

score_cv = cross_val_score(model_clf, x_train, y_train, cv=10, scoring='roc_auc')
print('Score Cross Validation : ', score_cv)
```

```
# membuat model dan evaluasi data dengan algoritma KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

model_knn = KNeighborsClassifier()
model_knn.fit(x_train, y_train)

score_cv = cross_val_score(model_knn, x_train, y_train, cv=10, scoring='roc_auc')
print('Score Cross Validation : ', score_cv.mean())
```

```
# membuat model dan evaluasi data dengan algoritma GaussianNB
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score

model_gnb = GaussianNB()
model_gnb.fit(x_train, y_train)

score_cv = cross_val_score(model_gnb, x_train, y_train, cv=10, scoring='roc_auc')
print('Score Cross Validation : ', score_cv.mean())
```

```
# membuat model dan evaluasi data dengan algoritma GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

model_svc = SVC()
model_svc.fit(x_train, y_train)

score_cv = cross_val_score(model_svc, x_train, y_train, cv=10, scoring='roc_auc')
print('Score Cross Validation : ', score_cv)
```
